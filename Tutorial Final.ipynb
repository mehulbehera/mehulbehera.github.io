```python
#The objective of this project is analyze how COVID-19 affected the mental health of college students
#For the first couple steps of this assignment we are going to have to import pandas
#This will allow use to import online data sets into dataframes and plot those dataframes
import pandas as pd
```

```python
#The COVID-19 pandemic has affected everyone around the world. Not only has it affected us dramatically physically,
#with millions dying, it also has affected the way our world functions. To combat the pandemic we were forced 
#into social isolation which has taken a toll on societal mental health. Furthermore, school has been conducted 
#online for the safety of the students. Today we are going to take a look at how these conditions affected the 
#mental health of these students
```

```python
# Step 1: Gathering data
# These are the studies we are going to use to perform the analysis. We want data which can help us see the 
# change between college students mental health before and after COVID-19. This is how we know COVID-19 made an 
# impact on their health

BeforeAfterCOVID = pd.read_excel ('DepressionBeforeAfterCOVID.xlsx')
DisorderStats = pd.read_excel ('CollegeStudentMentalDisorderStat.xlsx')
DepressionStats = pd.read_excel ('CollegeStudentsDepression.xlsx')
SourceOfHelp = pd.read_excel ('SourceOfHelp.xlsx')
```

```python
#Step 2: Data Visualization
#Now that we have data we want to place in a digestable form so the audience can understand with 
#without having to run through the details and numbers
```

```python
print(BeforeAfterCOVID)

#https://www.statista.com/statistics/1184765/depression-level-increases-in-adults-due-to-covid-by-age-worldwide/
#This report contains data on depression rates before and after covid

#Because this data is presented in a manner where we want to compare the before and after the best options of 
#presentation are with a bar or line graph

ax = BeforeAfterCOVID.plot.bar(x = 'Age', rot =0, 
                               color={"No more depressed than before": "blue", "More depressed": "red"},figsize=(20,10))

#As we. can after. plot the bar plot, the data is a lot easier to register.
#Clearly. 18-24. year olds, mostly college students have been impacted the most by COVID with an alarming 35% increase
#in depression. Now that we know college are getting the brunt of the mental health problems, lets try to dive deeper
#and get more information on the situation

```

```python

DisorderStats.index = DisorderStats['Disorder']
#https://www.statista.com/statistics/1126287/percentage-of-college-students-with-mental-disorders-us/
#This data set tells us the percentage of college diagnosed with mental disorders during COVID-19

DisorderStats.plot.pie(
    y='Percentage of college students with diagnosed mental disorders in the United States in 2020 to 2021, by disorder', figsize=(100,10))

#From the pie change we can see the break down of mental disorders diagnosed from 2020 to 2021
#This gives us insight into how many college students specifically had depression during the COVID pandemic.
```

```python
#https://www.statista.com/statistics/1126279/percentage-of-college-students-with-depression-us/
#This data set tells us the severity of depression in college students during COVID-19

DepressionStats.index = DepressionStats['Depression Type']

DepressionStats.plot.pie(
    y='Percentage of college students with symptoms of depression in the United States in 2020 to 2021', figsize=(100,10))

#The pie chart of this data set allows us to see what categories of depression students have, and most importantly,
#how many students are suffering from severe depression
```

```python

#https://www.statista.com/statistics/1126746/informal-help-seeking-mental-health-college-students-us/
#This data tells us where students went to get help to deal with mental illness during the pandemic

SourceOfHelp.index = SourceOfHelp['Support group']
ax = SourceOfHelp.plot.bar(rot =0,figsize=(35,20))

#This bar chart allows us to see where students choose to get their support from
#This could be very insightful in terms of trying to combat the growing mental illness issue.
```

```python
#Before, we move on to the next step, something important to take note of is what the combination of these data sets
#tell us. There is a significant increase in depression amongst college students during the pandemic to the point
#where almost 40% of the population suffer from some form of anxiety or depression. Whats even more alarming
#is that a major chunk of those sufferening do not have a support system they feel comfortable using for aid.
# Social isolation may play a role in whether or not people choose to be silent about their depression, because
#many people have grown apart from friends and family.
```
